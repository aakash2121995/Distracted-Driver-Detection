```python
import pandas as pd
import numpy as np
import os
from shutil import copyfile,copy2
```

```python
csv = pd.read_csv('driver_imgs_list.csv')
```

```python
csv
```

```python
csv = csv.as_matrix()
```

```python
csv[0][1]
```

```python
i = 1
j = 1
os.mkdir('Training Data/')
os.mkdir('Test Data/')
os.mkdir('Validation Data/')

for row in csv:
    if(not os.path.isdir('./Training Data/' + row[1])):
        os.mkdir('./Training Data/' + row[1])
    if(not os.path.isdir('./Test Data/' + row[1])):
        os.mkdir('./Test Data/' + row[1])
    if(not os.path.isdir('./Validation Data/' + row[1])):
        os.mkdir('./Validation Data/' + row[1])
    if i%4 == 0:
        copy2("train/" + row[1] + '/' + row[2],'./Test Data/' + row[1] + '/')
    else:
        if j%10 == 0:
            copy2("train/" + row[1] + '/' + row[2],'./Validation Data/' + row[1] + '/')
        else:
            copy2("train/" + row[1] + '/' + row[2],'./Training Data/' + row[1] + '/')
        j = j+1
    i = i+1
    
    if i%500 == 0:
        print(i)
        
    
#     brea
```
